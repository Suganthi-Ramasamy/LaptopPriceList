# Filter Laptops by brand

```python
import pandas as pd
import numpy as np
pricelist = pd.read_table('/content/Laptop-PriceList-New.csv', sep=",")
pricelist.head()
```

# Filter by RAM and SSD  
Extract laptops that have at least 8 GB of RAM and an SSD of 256 GB or more.

```python
cols = ['brand','ram_gb', 'ssd_gb']
pricelist[(pricelist.ram_gb >= 8 ) & (pricelist.ssd_gb >= 256)][cols]
```

# Sort by Price
Sort the filtered laptops by price in ascending and descending order

```python
#Sorting by price in ascending order
pricelist.sort_values(by='Price', ascending=True).head()
```

```python
#Sorting by price in descending order
pricelist.sort_values(by='Price', ascending=False).head()
```

# Filter Touchscreen Laptops
Extract laptops that have a touchscreen feature


```python
pricelist[pricelist['Touchscreen'] == 'Yes'].head()
```

# Combine Filters
Extract laptops that meet multiple criteria, e.g., brand "HP", at least 8 GB RAM, and touchscreen.

```python
col1 = ['brand','ram_gb', 'Touchscreen']
pricelist[(pricelist.brand == 'HP') & (pricelist.ram_gb >= 8) & (pricelist.Touchscreen == 'Yes')][col1]
```
